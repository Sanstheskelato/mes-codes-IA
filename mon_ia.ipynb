```python
import tensorflow as tf
```

## inputs

```python
import numpy as np

x_train = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=np.float32)
y_train = np.array([2, 3, 6, 6, 10, 9, 14, 12], dtype=np.float32)

x_train_puissance = np.stack([np.sin(x_train), np.cos(x_train),np.sin(2*x_train), np.cos(2*x_train)], axis=1)
x_train_puissance.shape
```

# foward propagation et back propagation

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='sigmoid', input_shape=(4,)),
    tf.keras.layers.Dense(128, activation='sigmoid'),
    tf.keras.layers.Dense(128, activation='sigmoid'),
    tf.keras.layers.Dense(1)
])


model.compile(optimizer="adam", loss='mean_squared_error')
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

model.fit(x_train_puissance, y_train, epochs=1000, verbose=1)

plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train_puissance), color='red')
plt.show()
```

# prediction

```python
x = int(input("donne un nombre faisant partie de la liste: "))
y = int(input("donne un nombre faisant partie de la liste: "))
z = int(input("donne un nombre faisant partie de la liste: "))
w = int(input("donne un nombre faisant partie de la liste: "))

x_test = np.array([x, y, z, w], dtype=np.float32)
x_test_cube = np.stack([np.sin(x_test), np.cos(x_test),np.sin(2*x_test), np.cos(2*x_test)], axis=1)
y_pred = model.predict(x_test_cube)
print("Prédictions :", y_pred)
```
